```python
data=data.fillna(method='ffill')
```

```python
y= data['class']
x= data.drop('class', axis=1)
```

```python
select = SelectKBest(score_func=f_regression)
```

```python
select.fit(x,y)
```

```python
score_col=pd.DataFrame(select.scores_,columns=['Score value'])
```
